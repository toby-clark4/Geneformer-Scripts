```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
kf0_dists = pd.read_csv('/mnt/scratchc/ghlab/in_silico_perturb_data/gene_embs/kf0/in_silico_overexpress_cnot1_1_dict_1Kbatch_euclidean_dists.csv', index_col=0)
kf1_dists = pd.read_csv('/mnt/scratchc/ghlab/in_silico_perturb_data/gene_embs/kf1/in_silico_overexpress_cnot1_1_dict_1Kbatch_euclidean_dists.csv', index_col=0)
kf2_dists = pd.read_csv('/mnt/scratchc/ghlab/in_silico_perturb_data/gene_embs/kf2/in_silico_overexpress_cnot1_1_dict_1Kbatch_euclidean_dists.csv', index_col=0)
all_dists = pd.DataFrame()
all_dists['kf0_dist'] = kf0_dists['euclidean_distance']
all_dists['kf1_dist'] = kf1_dists['euclidean_distance']
all_dists['kf2_dist'] = kf2_dists['euclidean_distance']
all_dists['mean_dist'] = all_dists.apply(np.mean, axis=1)
all_dists['token'] = kf0_dists['token']
all_dists.to_csv('/mnt/scratchc/ghlab/in_silico_perturb_data/gene_embs/avg_dists_cnot1_oex_cell_1.csv')
```

```python
tf_list = ['ENSG00000183072',
           'ENSG00000106459',
           'ENSG00000113196',
           'ENSG00000112658',
           'ENSG00000135111',
           'ENSG00000089225',
           'ENSG00000164532',
           'ENSG00000136574',
           'ENSG00000130700',
           'ENSG00000141448',
           'ENSG00000068305']

control_list = ["ENSG00000075624",
                "ENSG00000111640",
                "ENSG00000166710",
                "ENSG00000165704",
                "ENSG00000087245",
                "ENSG00000089157",
                "ENSG00000136560",
                "ENSG00000150991",
                "ENSG00000078808",
                "ENSG00000102144"]

tf_dist_list=[]
ctrl_dist_list=[]
for gene in tf_list:
    print(gene)
    df = pd.read_csv(f'/mnt/scratchc/ghlab/in_silico_perturb_data/in_silico_pt_embs/heart_proof_of_concept/in_silico_delete_{gene}_dict_1Kbatch_euclidean_dists.csv', index_col=0)
    tf_dist_list.append(df['euclidean_distance'].median())
for gene in control_list:
    df = pd.read_csv(f'/mnt/scratchc/ghlab/in_silico_perturb_data/in_silico_pt_embs/heart_proof_of_concept/in_silico_delete_{gene}_dict_1Kbatch_euclidean_dists.csv', index_col=0)
    ctrl_dist_list.append(df['euclidean_distance'].median())
                          
combined_df = pd.DataFrame({'distances':tf_dist_list+ctrl_dist_list,'label':['heart' for x in range(len(tf_dist_list))] + ['ctrl' for x in range(len(ctrl_dist_list))]})
```

```python
combined_df.to_csv('/mnt/scratchc/ghlab/in_silico_perturb_data/in_silico_pt_embs/heart_proof_of_concept/combined_df_median_dists.csv')
```
