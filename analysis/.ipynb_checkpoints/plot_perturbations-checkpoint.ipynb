```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats
import pickle
```

```python
def plot_violin(data, title, xlabel, ylabel, xticklabels, palette, show_p=True, show_n=True, figsize=[6.4, 4.8], filepath=None, ax=None):
    """
    Create a box-and-whisker plot with significance bars.
    """
    if ax is None:
        fig, ax = plt.subplots(figsize=figsize)
    else:
        fig = ax.figure
 
    bp = sns.violinplot(data, widths=0.8,palette=palette, inner_kws={'width':20,'native_scale':False})
    
    # Graph title
    ax.set_title(title, fontsize=14)
    # Label x-axis
    ax.set_xlabel(xlabel)
    # Label y-axis
    ax.set_ylabel(ylabel)
    # Label x-axis ticks
    # Hide x-axis major ticks
    ax.tick_params(axis="x", which="major", length=0)
    # Show x-axis minor ticks
    xticks = [0.5] + [x + 0.5 for x in ax.get_xticks()]
    ax.set_xticks(xticks, minor=True)
    # Clean up the appearance
    ax.tick_params(axis="x", which="minor", length=3, width=1)
    ax.set_xticklabels(xticklabels, fontsize=10)
     
    # Check for statistical significance
    significant_combinations = []
    # Check from the outside pairs of boxes inwards
    ls = list(range(1, len(data) + 1))
    combinations = [(ls[x], ls[x + y]) for y in reversed(ls) for x in range((len(ls) - y))]
    for c in combinations:
        data1 = data[c[0] - 1]
        data2 = data[c[1] - 1]
        # Significance
        U, p = stats.mannwhitneyu(data1, data2, alternative="two-sided")
        #U, p = stats.wilcoxon(data1, data2, alternative="two-sided")
        if p < 0.05:
            significant_combinations.append([c, p])
            # Find which series is better, using average rank
            median1 = np.median(data1)
            median2 = np.median(data2)
            
            if median1 > median2:
                res_op = f"\u227b, {np.abs(1-median1/median2)*100:.0f}%"
            elif median1 < median2:
                res_op = f"\u227a, {np.abs(1-median1/median2)*100:.0f}%"
            else:
                res_op = "\u2248"
             
    # Get info about y-axis
    bottom, top = ax.get_ylim()
    yrange = top - bottom
 
    # Significance bars
    for i, significant_combination in enumerate(significant_combinations):
        # Columns corresponding to the datasets of interest
        x1 = significant_combination[0][0]-1
        x2 = significant_combination[0][1]-1
        # What level is this bar among the bars above the plot?
        level = len(significant_combinations) - i
        # Plot the bar
        bar_height = (yrange * 0.08 * level) + top
        bar_tips = bar_height - (yrange * 0.02)
        ax.plot(
            [x1, x1, x2, x2],
            [bar_tips, bar_height, bar_height, bar_tips], lw=1, c="gray")
        # Significance level
        p = significant_combination[1]
        if p < 0.001:
            sig_symbol = "***"
            sig_text = fr"$p < 0.001$"
        elif p < 0.01:
            sig_symbol = "**"
            sig_text = fr"$p=${p:.3f}"
        elif p < 0.05:
            sig_symbol = "*"
            sig_text = fr"$p=${p:.2f}"
        
        sig_text = f"{sig_text}"
        
        text_height = bar_height + (yrange * 0.01)
 
        if show_p:
            ax.text((x1 + x2) * 0.5, text_height, sig_text, ha="center", c="k", size='small')
        else:
            ax.text((x1 + x2) * 0.5, text_height, sig_symbol, ha="center", c="k", size="small")
 
    # Adjust y-axis
    bottom, top = ax.get_ylim()
    yrange = top - bottom
    ax.set_ylim(bottom - 0.02 * yrange, top)
 
    # Annotate sample size below each box: if show_n is passed
    if show_n:
        for i, dataset in enumerate(data):
            sample_size = len(dataset)
            ax.text(i, bottom, fr"$n = {sample_size}$", ha="center", size="medium", color="blue")
 
    if filepath:
        plt.savefig(filepath, dpi=300, bbox_inches="tight")
 
    return fig, ax
```

```python
def plot_box_and_whisker(data, title, xlabel, ylabel, xticklabels, palette, show_p=True, show_n=True, figsize=[6.4, 4.8], filepath=None, ax=None):
    """
    Create a box-and-whisker plot with significance bars.
    """
    if ax is None:
        fig, ax = plt.subplots(figsize=figsize)
    else:
        fig = ax.figure
 
    bp = sns.boxplot(data, palette=palette, showfliers=False, width=0.3)
    # Graph title
    ax.set_title(title, fontsize=10)
    # Label x-axis
    ax.set_xlabel(xlabel)
    # Label y-axis
    ax.set_ylabel(ylabel)
    # Label x-axis ticks
    ax.set_xticklabels(xticklabels, fontsize=10)
    # Hide x-axis major ticks
    ax.tick_params(axis="x", which="major", length=0)
    # Show x-axis minor ticks
    xticks = [0.5] + [x + 0.5 for x in ax.get_xticks()]
    ax.set_xticks(xticks, minor=True)
    # Clean up the appearance
    ax.tick_params(axis="x", which="minor", length=3, width=1)
 
 # Check for statistical significance
    significant_combinations = []
    # Check from the outside pairs of boxes inwards
    ls = list(range(1, len(data) + 1))
    combinations = [(ls[x], ls[x + y]) for y in reversed(ls) for x in range((len(ls) - y))]
    print(combinations)
    for c in combinations:
        data1 = data[c[0] - 1]
        data2 = data[c[1] - 1]
        # Significance
        U, p = stats.mannwhitneyu(data1, data2, alternative="two-sided")
        print(U, p)
        #U, p = stats.wilcoxon(data1, data2, alternative="two-sided")
        if p < 0.05:
            significant_combinations.append([c, p])
            # Find which series is better, using average rank
            median1 = np.median(data1)
            median2 = np.median(data2)
            
            if median1 > median2:
                res_op = f"\u227b, {np.abs(1-median1/median2)*100:.0f}%"
            elif median1 < median2:
                res_op = f"\u227a, {np.abs(1-median1/median2)*100:.0f}%"
            else:
                res_op = "\u2248"
             
    # Get info about y-axis
    bottom, top = ax.get_ylim()
    yrange = top - bottom
    # Significance bars
    print(significant_combinations)
    for i, significant_combination in enumerate(significant_combinations):
        # Columns corresponding to the datasets of interest
        x1 = significant_combination[0][0]-1
        x2 = significant_combination[0][1]-1
        # What level is this bar among the bars above the plot?
        level = len(significant_combinations) - i
        # Plot the bar
        bar_height = (yrange * 0.08 * level) + top
        bar_tips = bar_height - (yrange * 0.02)
        ax.plot(
            [x1, x1, x2, x2],
            [bar_tips, bar_height, bar_height, bar_tips], lw=1, c="gray")
        print(bar_height)
        # Significance level
        p = significant_combination[1]
        if p < 0.001:
            sig_symbol = "***"
            sig_text = fr"$p < 0.001$"
        elif p < 0.01:
            sig_symbol = "**"
            sig_text = fr"$p=${p:.3f}"
        elif p < 0.05:
            sig_symbol = "*"
            sig_text = fr"$p=${p:.2f}"
        
        sig_text = f"{sig_text}"
        
        text_height = bar_height + (yrange * 0.01)
 
        if show_p:
            ax.text((x1 + x2) * 0.5, text_height, sig_text, ha="center", c="k", size='small')
        else:
            ax.text((x1 + x2) * 0.5, text_height, sig_symbol, ha="center", c="k", size="small")
 
    # Adjust y-axis
    bottom, top = ax.get_ylim()
    yrange = top - bottom
    ax.set_ylim(bottom - 0.02 * yrange, top)
 
    # Annotate sample size below each box: if show_n is passed
    if show_n:
        for i, dataset in enumerate(data):
            sample_size = len(dataset)
            ax.text(i, bottom, fr"$n = {sample_size}$", ha="center", size="medium", color="blue")
 
    if filepath:
        plt.savefig(filepath, dpi=300, bbox_inches="tight")
 
    return fig, ax
```

```python
dir_list = ['cnot1', 'eif4a1','dmd','tead4','actb']
dist_df = pd.DataFrame()
for direc in dir_list:
    emb = pd.read_csv(f'/mnt/scratchc/ghlab/in_silico_perturb_data/in_silico_pt_embs/{direc}/in_silico_overexpress_{direc}_dict_1Kbatch_euclidean_dists.csv', index_col=0)
    emb = emb.rename(columns={'euclidean_distance':direc.upper()})
    dists_per_cell = emb.transpose().iloc[0,:]
    dists_per_cell = dists_per_cell.rename({'euclidean_distance':direc})
    dist_df = pd.concat([dist_df, dists_per_cell], axis=1)
dist_df
```
